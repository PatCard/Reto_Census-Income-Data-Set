```python
#Importamos las librerias principales
import pandas as pd
```

```python
#Cargamos dataset a utilizar
ingresos = pd.read_csv("dataset/adult.data", sep = ",")
```

```python
ingresos.head(10)
```

```python
#Analizamos el shape del objeto
print(ingresos.shape)
```

```python
#Como podemos observar vamos a renombrar las columnas de manera correcta
col_names = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'salary']
ingresos.columns = col_names
```

```python
ingresos.head(10)
```

```python
#Visualizamos los tipos de datos
ingresos.dtypes
```

```python
# Separar las características de la variable objetivo
X = ingresos.drop(['salary'], axis=1)
y = ingresos['salary']
```

```python
from category_encoders import OrdinalEncoder
# Transformar las variables categóricas con OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
X = ordinal_encoder.fit_transform(X)
```

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Se crea el modelo de Random Forest
from sklearn.ensemble import RandomForestClassifier

random_forest = RandomForestClassifier(random_state=42)
```

```python
#Entrenamiento
random_forest.fit(X_train,y_train)
```

```python
#Calculo de las predicciones en Train y Test
y_train_pred = random_forest.predict(X_train)
y_test_pred = random_forest.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
# Calcular la precisión del modelo
#Calculo el accuracy en Train
train_accuracy = accuracy_score(y_train, y_train_pred)

#Calculo el accuracy en Test
test_accuracy = accuracy_score(y_test, y_test_pred)

print('El accuracy en train es:', train_accuracy)
print('El accuracy en test es:', test_accuracy)
```

```python
# Probabilidad de obtener mas us$50000
probabilidad = random_forest.predict_proba(X_test)

prob_50000 = probabilidad[:, 1]

# Calculo de la probabilidad promedio
probabilidad_promedio = prob_50000.mean() * 100

print(f"La probabilidad promedio de que alguien tenga un salario superior a US$50000 es de {probabilidad_promedio:.2f}%")
```
